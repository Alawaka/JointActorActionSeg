```python
import sys
sys.path.append('../')
from dataIO import A2D
from utils import *
import matplotlib.pyplot as plt
import matplotlib.patches as patches
%matplotlib inline
import numpy as np
import os
from os.path import join as opj
import skimage.io
```

```python
def apply_mask(image, mask, color, alpha=0.5):
    """Apply the given mask to the image.
       color RGB in range(0, 256)
    """
    for c in range(3):
        image[:, :, c] = np.where(mask != 0,
                                  image[:, :, c] *
                                  (1 - alpha) + alpha * color[c],
                                  image[:, :, c])
    return image
```

```python
def display_instances(image, masks, colors, class_names, fname,
                      scores=None, title="", boxes=None, 
                      figsize=20, ax=None, return_flag=False):
    """
    boxes: [num_instance, (y1, x1, y2, x2, class_id)] in image coordinates.
    masks: [num_instances, height, width]
    class_ids: [num_instances]
    class_names: list of class names of the dataset
    scores: (optional) confidence scores for each box
    figsize: (optional) the size of the image.
    """
    # Number of instances
    N = masks.shape[0]
    if not N:
        print("\n*** No instances to display *** \n")
    else:
        assert masks.shape[0] == colors.shape[0]
        if boxes is not None:
            assert masks.shape[0] == boxes.shape[0]

    height, width = image.shape[:2]
    if not ax:
        fig = plt.figure()
        fig.set_size_inches(figsize, figsize*height*1./width, forward=False)
        ax = plt.Axes(fig, [0., 0., 1., 1.])
        ax.set_axis_off()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        fig.add_axes(ax)

    masked_image = image.astype(np.uint32).copy()
    for i in range(N):
        color = colors[i]
        
        # Mask
        mask = masks[i, :, :]
        masked_image = apply_mask(masked_image, mask, color)
        
        # Bounding box
        if boxes is not None:
            x1, y1, w, h = boxes[i]
            y2 = y1 + h
            x2 = x1 + w
            p = patches.Rectangle((x1, y1), x2 - x1, y2 - y1, linewidth=2,
                                  alpha=0.7, linestyle="dashed",
                                  edgecolor=color*1./255, facecolor='none')
            ax.add_patch(p)
        
        # Label
        score = scores[i] if scores is not None else None
        label = labels[i]
        caption = "{} {:.3f}".format(label, score) if score else label
        ax.text(10, 10 + 20*i, caption,
                color=color*1./255, size=20, backgroundcolor="none")

    ax.imshow(masked_image.astype(np.uint8))
    plt.savefig(fname=fname, bbox_inches='tight', pad_inches=0) 
    plt.close()
    if return_flag:
        return masked_image.astype(np.uint8)
```

```python
a2d = A2D()
col2lab = a2d.getColorLabelLookupTable()
annotation_list = a2d.getAnnos()
img_list = []
for anno in annotation_list:
    splits = anno.split('/')
    img_list.append(opj(a2d.img_dir, splits[-2], splits[-1]))
```

```python
mask_dir = opj(a2d.root_dir, 'MaskedImages')
if not os.path.exists(mask_dir):
    os.mkdir(mask_dir)
vids = a2d.getVids()
for vid in vids:
    if not os.path.exists(opj(mask_dir, vid)):
        os.mkdir(opj(mask_dir, vid))
```
